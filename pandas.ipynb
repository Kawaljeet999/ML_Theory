```python
df = pd.DataFrame(
    {
    "name":['Afread','Batman','Afread'],
    "toy":[np.nan,'Batmobile','Bullwhip'],
    "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]
    }
)
```

```python
df.head()
```

```python
df.dropna() # where ever there is na drop it no matter it is column or row
```

```python
df.dropna(how='all') # it mean all values in a particular  row until all values  is na  we cant delete any row 
```

```python
df.dropna(how='all',axis=1) # it mean all values in a particular  column until all values  is na  we cant delete any  column
```

```python
df.drop_duplicates(subset=['name'],keep='first') # remove all occurencess except  first one
```

```python
df.drop_duplicates(subset=['name'],keep='last') # remove all occurencess except last one
```

```python
df.drop_duplicates(subset=['name'],keep=False) # remove all occurencess 
```

```python
df.shape
```

```python
df.info() # give info of dataframe
```

```python
df['toy'].value_counts(dropna=False) # Dont Remove na and count it it give only unique value
```

```python
df['toy'].value_counts(dropna=True) # Remove na and dont count it it give only unique value
```

```python
df.notnull()
```
