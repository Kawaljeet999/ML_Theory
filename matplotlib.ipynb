```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4] 
y = [0,2,4,6,8]

# 1 inch = 2.5cm
# Resize our Graph
plt.figure(figsize=(5,3), dpi=100) # dpi is pixels per inch and 5*100 =500 inch lenght/x axis
# and 3*100 = 300 inch is width/breath/y axis pixels per inch
# always get dpi high for good printout and chlarity

# Line 1
# Big Hand Notation/ Keyword argument notation
# plt.plot(x,y, label='2x', color='red', linewidth=2, linestyle='--', marker='.', markersize=10, markeredgecolor='Blue') 
# x & y must be 1D array or a scalar this plot used to plot the graph y mean ywllow and #ababa show different
# colours marker mean add anything in line

# Short Hand Notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b.--', label='2x')  # 'b,--' here b mean blue and . mean point and -- mean linestyle

# Line 2
# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)
print(x2)
# plot part of graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2') # here x2[:6] mean take first 6 elements in  numpy array in x axis and x2[:6]**2 mean to do 
# square of these elements in yaxis

# Pot remaning of graph
plt.plot(x2[5:], x2[5:]**2, 'r--') # here we take after 5 elements in numpy array do same thing

plt.title("Our First Graph", fontdict={"fontname":'Comic Sans MS',"fontsize":20}) 
plt.xlabel("X Axis!")
plt.ylabel("Y Axis!")

# X,Y tickmarks (scale of your graph)
# plt.xticks([0,1,2,3,4])  # change ticks of x axis 
# plt.yticks([0,2,4,6,8,10]) # change ticks of y axis

plt.legend() # it takes label implicity until not defined explicitly here 2x is a label and X^2

plt.savefig('mygraph', dpi=300) # it save graph with dpi 300 mean 1500 by 900 inches in lenght by breath respectively

plt.show() # this is uded to show the graph
```

### Pie chart #2

```python
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight] 

# for x in fifa.Weight: Iterates through each element (x) in the Weight column of fifa.
# type(x)==str: Checks if the current element (x) is a string.
# x.strip('lbs'): If x is a string, removes trailing "lbs" using strip('lbs').
# else x: If x is not a string, keeps it unchanged.
# here the value of weight is string and we cant convert it into pie chart so we need to use type conversion int(x.strip('lbs'))

fifa.Weight
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa[fifa.Weight >= 200].count().iloc[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4) # here we only give non zero values to these labels that are cut apart 
# explode cut apart the pie chart to make it more visible

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%') # pctdistance is percentage distance from center 

plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(7,9), dpi=100)
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

labels = ['FC Barcelona', "Real Madrid", "New England Revolution"]

bp = plt.boxplot([barcelona, madrid,revs], labels=labels, patch_artist=True, medianprops={'linewidth':2}) # here labels = labels is a keyword - value pair 
# if we set patch_artist = True then only we change face color
# here medianprops  change yellow linw width

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.xlabel('FIFA Club')


plt.show()
```
